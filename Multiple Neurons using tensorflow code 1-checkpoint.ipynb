```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler,LabelEncoder
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
tip=sns.load_dataset('tips')
```

```python
tip
```

```python
le=LabelEncoder()
```

```python
tip['sex']=le.fit_transform(tip['sex'])
tip['smoker']=le.fit_transform(tip['smoker'])
tip['day']=le.fit_transform(tip['day'])
tip['time']=le.fit_transform(tip['time'])
```

```python
##convert tips to a binary variable 1 if the tips is above s2 else 0
```

```python
tip['tip']=(tip['tip'] > 2).astype(int)
```

```python
x=tip.drop('tip',axis=1)
y=tip['tip']
```

```python
##scaler the data
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(x)
```

```python
### train test split the data
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=42)
```

```python
#create a model for neural network
model=tf.keras.Sequential([
    tf.keras.layers.Dense(16,activation='relu',input_shape=[X_train.shape[1]]), # input layers
    tf.keras.layers.Dense(8,activation='relu'), # hidden layer
    tf.keras.layers.Dense(1,activation='sigmoid') # output layer
])
```

```python
X_train.shape[1]
```

```python
## compile the model
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
## batch size ( in one iteration how many  samples processes do it
## train the model
model.fit(X_train,y_train,epochs=5,batch_size=32,verbose=1)
```

```python
## evaluate the model l
loss,accuracy=model.evaluate(X_test,y_test)
print('Test loss : ' , loss)
print('Test accuracy : ' , accuracy)
```
